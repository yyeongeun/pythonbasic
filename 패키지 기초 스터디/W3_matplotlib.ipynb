## 이중축 지정

같은 plot()으로 x축은 공유하되, y축의 스케일을 다르게 하고 싶을 때는 `twinx()` 를 이용합니다. 

1번째 plot()은 차트 왼쪽에 y축 스케일이, 2번째 plot()은 우측에 y축 스케일이 보이는 것을 확인하실 수 있습니다.

```python
plt.rcParams['axes.unicode_minus'] = False 
#마이너스 부호를 출력할 수 있게 하는 코드입니다. 만약 이 코드를 넣지 않고 실행했을 때 마이너스 부호가 깨진다면 이 코드를 써줍시다.

x = np.arange(1, 6)
y1 = - x -3
y2 = x *2

fig, ax1 = plt.subplots()
ax1.plot(x, y1, color='green')

ax2 = ax1.twinx()
ax2.plot(x, y2, color='deeppink')

plt.show()
```

## 범례 그리기
`legend()`를 이용하여 각 데이터가 무엇인지 설명해주는 범례를 넣을 수도 있습니다.

* legend() : 차트에 각 데이터의 설명 추가

```python
x = np.linspace(0, 2*np.pi, 50)

plt.plot(np.sin(x))
plt.plot(np.cos(x))
plt.legend(['sin', 'cos'])

plt.show()
```

범례의 위치는 수동으로 설정할 수 있습니다. 
legend()함수의 loc패러미터를 통해 범례의 위치를 문자열, 혹은 숫자 코드를 통해 지정가능 합니다.


![legend](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FED4sP%2FbtqFrUUtElB%2FoSjYDFsBKKaRqQksCCeHJK%2Fimg.png)

0은 디폴트 값으로, 범례를 그래프의 최적의 위치에 표시합니다.

또한, ncol 패러미터를 통해 범례에 표시될 텍스트 열 개수를 지정할 수 있습니다. 디폴트 값은 1로, 1개의 열로 표시됩니다.

```python
x = np.linspace(0, 2*np.pi, 50)

plt.plot(np.sin(x))
plt.plot(np.cos(x))
plt.legend(['sin', 'cos'],loc=10,ncol=2)

plt.show()
```

이외에도, legend()함수는 범례 박스 테두리나 그림자 설정 등 다양한 기능을 가지고 있고, 더 자세한 애용은 matplotlib의 공식 홈페이지에서 확인할 수 있습니다.

## subplot

하나의 차트에 여러 데이터를 중복해서 그리는 것이 아니라, 하나의 이미지에서 여러 부분으로 나누어 각각의 차트를 만들 수도 있습니다. 이때 `subplot()`을 이용합니다. 

세 가지(nrows, ncols, index)를 입력을 받으면 nrows는 행의 개수, ncols는 열의 개수, index는 보여줄 위치로 1부터 지정하고 왼쪽 위부터 오른쪽 아래로 하나씩 지정됩니다. 

세 개의 인수 (2,2,3)을 넣고 싶다면 숫자 223을 넣어도 동작합니다.

* subplot(nrows, ncols, index): 행의 개수, 열의 개수, 차트 위치 순으로 입력
* subplot(###): 3자리 숫자로, 행/열/위치를 한번에 설정

```python
y1 = npr.normal(size = 50).cumsum()
y2 = npr.normal(size = 100).cumsum()
y3 = npr.normal(size = 50).cumsum()
y4 = npr.normal(size = 100).cumsum()

#2개 행, 2개 열로 구간을 나누고, 1번째 영역(왼쪽 위)에 플롯을 만들어라.
plt.subplot(2, 2, 1)
plt.plot(y1)

#2개 행, 2개 열로 구간을 나누고, 2번째 영역(오른쪽 위)에 플롯을 만들어라.'
plt.subplot(2, 2, 2)
plt.plot(y2)

#2개 행, 2개 열로 구간을 나누고, 3번째 영역(왼쪽 아래)에 플롯을 만들어라.
plt.subplot(2, 2, 3)
plt.plot(y3)

#2개 행, 2개 열로 구간을 나누고, 4번째 영역(오른쪽 아래)에 플롯을 만들어라.
plt.subplot(2, 2, 4)
plt.plot(y4)

plt.show()
```

## 어노테이션(annotation)
차트 안에 화살표를 포함한 부가 설명 문자 열을 넣을 수 있습니다. 이때 `annotation()`을 사용해줍니다.
- `plt.annotate('문자열', xy, xytext, arrowprops)`


문자열은 어노테이션에서 나타낼 문자열, xy는 화살표가 가르키는 점의 위치, xytext는 문자열이 출력될 위치, arrowprops는 화살표의 속성으로 칼라 등을 정의

```python
x = np.arange(1,10)
y = x*5

plt.plot(x,y)
plt.annotate('annotate',xy=(2,10),xytext=(5,20),arrowprops={'color':'green'})
plt.show()
```

## Errorbar 표기
plt.plot() 대신 plt.errorbar()를 사용하면 Line Plot의 각 포인트에 errorbar를 넣을 수 있습니다.

.errorbar(x, y, yerr, xerr): plot(x,y) 내부 각 포인트에 yerr크기의 세로선, xerr 크기의 가로선

만약 errorbar() 안에 입력 값으로 uplims, lolims 값을 설정해주면 에러바를 화살표 형식으로 표기할 수도 있습니다.

uplims : 아래로 향하는 화살표
lolims : 위로 향하는 화살표

```python
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)
xerr = np.linspace(0, 1, 10) / 5
yerr = np.linspace(1, 0, 10) / 5

plt.errorbar(x, y, yerr, xerr, 
             label = '디폴트로 설정된 에러바')

plt.errorbar(x, y + 1, yerr=yerr, uplims=True, 
             label='upilm 아래방향 화살표')

plt.errorbar(x, y + 2, yerr=yerr, uplims=[True, False] * 5, lolims=[False, True] * 5, 
             label='uplim lolim 둘다 쓰면')

plt.title('에러바의 종류')
plt.legend()
plt.show()
```

# 만든 플롯 저장하기

코드를 전부 작성하고, 맨 긑에 plt.savefig('이미지이름.형식', dpi=해상도)를 써주면 만들어준 플롯이 저장됩니다. 

저장경로는 이 노트북 파일이 있는 곳!

```python
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)
xerr = np.linspace(0, 1, 10) / 5
yerr = np.linspace(1, 0, 10) / 5

plt.errorbar(x, y, yerr, xerr, 
             label = 'default')

plt.errorbar(x, y + 1, yerr=yerr, uplims=True, 
             label='upilm')

plt.errorbar(x, y + 2, yerr=yerr, uplims=[True, False] * 5, lolims=[False, True] * 5, 
             label='uplim lolim')

plt.title('errorbar')
plt.legend()
plt.show()
plt.savefig('errorbar.png', dpi=200)
```

# histogram

plt.hist()를 통하여 히스토그램을 그릴 수 있습니다. 히스토그램은 연속적인 데이터를 막대그래프로 나타냅니다.

```python
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight)
plt.title('weight')
plt.xlabel('kg')
plt.ylabel('count')

plt.show()
```

```python
#히스토그램 응용

#bins 패러미터: 히스토그램의 가로축 구간 개수 지정. 디폴트는 10입니다.
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight, bins=20, label='bins=20')
plt.title('weight')
plt.xlabel('kg')
plt.ylabel('count')
plt.legend()
plt.show()
```

```python
#누적 히스토그램
#cumulative 패러미터를 통해 누적 히스토그램 작성
#bool 자료형을 입력받고, true로 설정하면 x 방향으로 진행할 수록 y축 값의 데이터가 축적되어 그래프를 그린다.
# 쉽게 풀어서 설명하자면, 아래 히스토그램에서, 오렌지색 블록이 x축 방향으로 진행할 수록 
# 바로 전의 오렌지색이 차례차례 더해진게 파란색의 높이라고 생각하면 됩니다..!

weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight, cumulative=True, label='cumulative=True')
plt.hist(weight,  cumulative=False, label='cumulative=False')
plt.legend(loc='upper left')

plt.show()
```

## 파이차트 기타 설정: 시작 각도와 방향, 파이 간 간격, 색상

`startangle`는 부채꼴이 그려지는 시작 각도를 설정합니다. 디폴트는 0도 (양의 방향 x축)로 설정되어 있습니다.

`counterclock=False`로 설정하면 시계 방향 순서로 부채꼴 영역이 표시됩니다.

# box plot, violin plot

박스 플롯 (Box plot) 또는 박스-위스커 플롯 (Box-Whisker plot)은 수치 데이터를 표현하는 하나의 방식입니다.

일반적으로 박스 플롯은 전체 데이터로부터 최소값, 제1사분위~제4사분위수, 최대값의 요약수치를 사용해 그립니다.


vert=false를 지정해주면 수평 방향의 박스플롯을 그릴 수 있습니다.

바이올린 플롯 (Violin plot)은 데이터의 분포와 범위를 한눈에 보기 쉽게 나타내는 그래프 형식입니다.

박스 플롯 (Box plot)과 비슷하지만 더 실제에 가까운 분포를 알 수 있다는 장점이 있습니다.

```python
#박스플롯
#1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c])
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

```python
#바이올린 플롯
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c])
ax.set_ylim(-10.0, 10.0)
ax.set_xticks([1, 2, 3])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

## 바이올린 플롯 최대/최소값, 중간값, 평균값 표시

showmeans 파라미터는 데이터 분포에서 평균값의 위치에 직선을 표시합니다. (Default: False)

showextrema는 최대/최소값에 직선을 표시합니다. (Default: True)

showmedians는 중간값에 직선을 표시합니다. (Default: False)

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

## 바이올린 플롯 색상 정하기

* violin[‘bodies’][n] set_facecolor() : n번째 바이올린 분포 영역의 색상을 지정합니다.

* violin[‘cbars’] set_edgecolor():중심을 표시하는 직선의 색상을 지정합니다.

* violin[‘cmaxes’] set_edgecolor():분포의 최대값을 표시하는 직선의 색상을 지정합니다.

* violin[‘cmins’] set_edgecolor():분포의 최소값을 표시하는 직선의 색상을 지정합니다.

* violin[‘cmeans’] set_edgecolor(): 분포의 평균을 표시하는 직선의 색상을 지정합니다.

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], positions=[2, 3, 4])
ax.set_ylim(-10.0, 10.0)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

#4. 각각 차트 색상 설정
ax.set_xticklabels(['A', 'B', 'C'])
violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

violin['cbars'].set_edgecolor('gray')
violin['cmaxes'].set_edgecolor('gray')
violin['cmins'].set_edgecolor('gray')
plt.show()
```

그렇지만 여기서는 임포트를 해보겠습니다.



### 위에서 다뤘던 스캐터 플롯을 3차원에 표시할 수 있습니다. 예제만 간단하게 살펴보도록 하죠.

기본 형식: `Axes3D.scatter(x, y, z)`

```python
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure()
ax = Axes3D(fig)

X = npr.normal(size = 100)
Y = npr.normal(size = 100)
Z = npr.normal(size = 100)

ax.scatter(X, Y, Z, marker='^') #따로 마커, 색상을 지정하지 않으면 가장 기본값으로 나옵니다. 마커의 경우 동그라미

ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_zlabel('Z values')
plt.title('3D scatter')
plt.show()
```

### 심화: plot_surface() 함수를 사용해서 2차원 히트맵을 3차원으로 표현할 수 있습니다.
기본 형태: `Axes3D.plor_surface(X,Y,Z)`

예시 이미지
![3d](https://codetorial.net/matplotlib/_images/3d_plot_surface_00.png)
plot_surface() 함수는 2차원 어레이 형태를 3차원 Surface로 표현합니다.




```python
from mpl_toolkits.mplot3d import Axes3D

#객체 선언
fig = plt.figure()
ax = Axes3D(fig)

#x,y 배열 지정
X = np.arange(-5,5,0.2)
Y = np.arange(-10,10,0.2)

# X,Y값을 직교좌표계로 변환. 3차원 표면 플롯 생성을 위해 필요.
X, Y = np.meshgrid(X, Y)     

# Z = (x제곱 + y제곱)의 근으로 설정해보았습니다.
Z = np.sqrt(X**2+Y**2)

# 3D 그래프 생성.  rstride는 row의 sample 추출 간격, cstride는 column의 추출간격.
three_d=ax.plot_surface(X, Y, Z, rstride=5, cstride=10, cmap='summer')  

#shrink는 컬러바의 크기, aspect는 컬러바의 종횡비를 지정합니다.
fig.colorbar(three_d,shrink=0.6, aspect=8)

plt.show()
```

이외에도 라인, 바 등 우리가 기존에 배웠던 플롯이나 차트를 3D 형식으로 만들어줄 수도 있고, 폴리곤, 와이어프레임, 컨투어, Tri-surface 등 다양한 3d 플롯을 만들 수 있습니다. 그렇지만 프로젝트를 수행할 때 이 정도의 스킬을 필요로 하는 경우는 많지 않으니, matplotlib에는 이런 것도 있다는 정도로만 알아둡시다.
